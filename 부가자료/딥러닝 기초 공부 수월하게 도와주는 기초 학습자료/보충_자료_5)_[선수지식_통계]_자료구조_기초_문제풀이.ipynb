4. <b>arr.clear()</b>: 리스트 내 모든 원소를 제거하는 함수로, 최악의 경우 시간 복잡도는 $O(1)$이다.
5. <b>arr.copy()</b>: 리스트를 복제하는 함수로, 최악의 경우 시간 복잡도는 $O(N)$이다.

#### <b>1. 힙 정렬의 시간 복잡도</b>

* <b>[문제 해설]</b> 힙(heap)에 원소를 삽입하거나 삭제할 때 시간 복잡도 $O(logN)$이 요구된다.
* <b>힙 정렬</b>: $N$개의 데이터를 힙에 삽입한 뒤에, $N$개의 데이터를 모두 힙에서 꺼낸다.
* 따라서, 전체 데이터를 삽입할 때 $O(NlogN)$, 삭제할 때 $O(NlogN)$이 요구된다.
* 이때, $O(2NlogN) = O(NlogN)$이므로, <b>정답</b>은 ③번이다.

#### <b>2. 그래프의 표현 방식</b>

* <b>[문제 해설]</b> 노드의 개수를 $V$, 간선의 개수를 $E$라고 하자. 이때, 다음과 같다.
1. 인접 리스트는 $O(V + E)$만큼의 메모리 공간이 필요하다.
2. 인접 행렬은 $O(V^2)$만큼의 메모리 공간이 필요하다.
* 문제에서 $V > E$라고 하였으므로, $V$가 1보다 크다면, $O(V^2) > O(2V) > O(V + E)$이다.
* <b>(정답)</b> 따라서 더 메모리 측면에서 효율적인 표현 방식은 <b>인접 리스트</b> 방식이다.

#### <b>3. 스택 연산</b>

  * <b>1 삽입</b>: [1]
  * <b>추출</b>: []
  * <b>5 삽입</b>: [5]

  * <b>9 삽입</b>: [9, 4, 3, 5]
* 따라서 <b>정답</b>은 ②번이다.

* 삽입 정렬(오름차순 정렬)에서는 자기보다 작은 원소를 만날 때까지(좌측 원소가 자기보다 작을 때까지) 좌측 원소와 위치를 교체한다.

#### <b>8. 이진 탐색 트리의 순회</b>

* <b>[문제 해설]</b> 순회 결과는 다음과 같다.
  * <b>전위 순회</b>: 4 → 2 → 1 → 8 → 7 → 12 
  * <b>중위 순회</b>: 1 → 2 → 4 → 7 → 8 → 12 
  * <b>후위 순회</b>: 1 → 2 → 7 → 12 → 8 → 4 

```python
from collections import deque


class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        self.root = self._insert(self.root, key)

    def _insert(self, node, key):
        if node is None:
            return Node(key)

        # 현재 노드의 key보다 작은 경우
        if node.key > key:
            node.left = self._insert(node.left, key)
        # 현재 노드의 key보다 큰 경우
        elif node.key < key:
            node.right = self._insert(node.right, key)

        return node

    def preorder(self):
        self._preorder(self.root)

    def _preorder(self, node):
        if node:
            print(node.key, end=' ')
            self._preorder(node.left)
            self._preorder(node.right)

    def inorder(self):
        self._inorder(self.root)

    def _inorder(self, node):
        if node:
            self._inorder(node.left)
            print(node.key, end=' ')
            self._inorder(node.right)

    def postorder(self):
        self._postorder(self.root)

    def _postorder(self, node):
        if node:
            self._postorder(node.left)
            self._postorder(node.right)
            print(node.key, end=' ')


arr = [4, 2, 8, 1, 7, 12]
bst = BinarySearchTree()
for x in arr:
    bst.insert(x)
print('전위 순회:', end=' ')
bst.preorder()
print('\n중위 순회:', end=' ')
bst.inorder()
print('\n후위 순회:', end=' ')
bst.postorder()
```
